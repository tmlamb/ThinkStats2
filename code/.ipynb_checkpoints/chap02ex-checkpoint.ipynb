Read the female respondent file and display the variables names.

```python
%matplotlib inline

import chap01soln
resp = chap01soln.ReadFemResp()
resp.columns
```

Make a histogram of <tt>totincr</tt> the total income for the respondent's family.  To interpret the codes see the [codebook](http://www.icpsr.umich.edu/nsfg6/Controller?displayPage=labelDetails&fileCode=MALE&section=R&subSec=7958&srtLabel=609776).

```python
import thinkstats2
hist = thinkstats2.Hist(resp.totincr)
```

Display the histogram.

```python
import thinkplot
thinkplot.Hist(hist, label='totincr')
thinkplot.Show()
```

Make a histogram of <tt>age_r</tt>, the respondent's age at the time of interview.

```python
ageHist = thinkstats2.Hist(resp.age_r)
```

Make a histogram of <tt>numfmhh</tt>, the number of people in the respondent's household.

```python
numfmhhHist = thinkstats2.Hist(resp.numfmhh)
```

Make a histogram of <tt>parity</tt>, the number children the respondent has borne.  How would you describe this distribution?

```python
parityHist = thinkstats2.Hist(resp.parity)
thinkplot.Hist(parityHist, label='parity')
thinkplot.Show()
```

Use Hist.Largest to find the largest values of <tt>parity</tt>.

```python
parityHist.Largest
```

Use <tt>totincr</tt> to select the respondents with the highest income.  Compute the distribution of <tt>parity</tt> for just the high income respondents.

```python
highincr = resp[resp.totincr == 14]

highParityHist = thinkstats2.Hist(highincr.parity)
thinkplot.Hist(highParityHist, label='parity')
thinkplot.Show()


```

Find the largest parities for high income respondents.

```python
highParityHist.Largest
```

Compare the mean <tt>parity</tt> for high income respondents and others.

```python
meanParityHighincr = highincr.parity.mean()

otherincr = resp[resp.totincr != 14]

meanParityOtherincr = otherincr.parity.mean()

print meanParityHighincr
```
